# Step 3: Data Analysis

The following questions can be answered by analysing the data using pandas

1. Identify cars with the minimum and maximum prices.
2. Count the number of cars with different body styles.
3. Calculate the average price of cars by body type and arrange them in descending order.
4. Calculate the average price of cars by different companies and arrange them in descending order.
5. Count the different car types based on the number of cylinders.

### Identify Cars with Minimum and Maximum Prices

Cars with Minimum Price

```python
min_price_car = df[df['price'] == df['price'].min()]
print(min_price_car)
```

Cars with Maximum Price

```python
max_price_car = df[df['price'] == df['price'].max()]
print(max_price_car)
```

### Count the number of cars with different body styles

Count Cars by Body Style

```python
body_style_count = df['body-style'].value_counts()
print(body_style_count)
```

### Calculate the average price of cars by body type and arrange them in descending order.

Average Price by Body Type

```python
avg_price_body_type = df.groupby('body-style')['price'].mean().sort_values(ascending=False)
print(avg_price_body_type)
```

### Calculate the average price of cars by different companies and arrange them in descending order.

Average Price by Company

```python
avg_price_company = df.groupby('company')['price'].mean().sort_values(ascending=False)
print(avg_price_company)
```

### Count the different car types based on the number of cylinders

Count Car Types by Number of Cylinders

```python
cnt_car_no_cylinder = df['num-of-cylinders'].value_counts()
print(cnt_car_no_cylinder)
```

## Step 4: Data Visualization

Let's explore the data by visualizing the following cases:
1. Company & price
2. Company & no of cylinders
3. Body type & price
4. Body type & cylinders
5. engine type & horse power

### Visualize the relationship between company and price:

```python
#bar chart
plt.figure(figsize=(10, 8))
plt.bar(df['company'], df['price'], color='purple')
plt.title('Company vs Price',fontsize=18,)
plt.xlabel('Company',fontsize=12)
plt.ylabel('Price',fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
#line chart
plt.figure(figsize=(10, 6))
plt.plot(df['company'], df['price'], color='yellow')
plt.title('Company vs Price',fontsize=18)
plt.xlabel('Company',fontsize=12)
plt.ylabel('Price',fontsize=12)
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.show()
```

The bar chart and the line chart illustrates the variation in prices across different companies. It shows that company named Mercedes-benz has highest price and Chevrolet has lowest price

### Visualize the relationship between company and the number of cylinders

```python
#bar chart
plt.figure(figsize=(10, 8))
company_cylinders = df.groupby(['company', 'num-of-cylinders']).size().unstack(fill_value=0)
company_cylinders.plot(kind='bar', stacked=True, figsize=(10, 8))
plt.title('Company vs Number of Cylinders',fontsize=18)
plt.xlabel('Company',fontsize=12)
plt.ylabel('Number of Cars',fontsize=12)
plt.xticks(rotation=90)
plt.show()
```

This stacked bar chart shows the distribution of cylinder counts among different car manufacturers. It highlights the cylinder configurations favored by each company, reflecting manufacturing preferences and trends.

### Visualize the relationship between body type and price

```python
#bar chart
plt.figure(figsize=(10, 6))
plt.bar(df['body-style'], df['price'],color='orange')
plt.title('Body Type vs Price',fontsize=18)
plt.xlabel('Body Style',fontsize=12)
plt.ylabel('Price',fontsize=12)
plt.show()
```

```python
#pie chart
body_style_avg_price = df.groupby('body-style')['price'].mean()
colors = ['darkblue', 'lightblue']
plt.figure(figsize=(12, 8))
plt.pie(body_style_avg_price, labels=body_style_avg_price.index, autopct='%1.1f%%', colors=colors, startangle=140)
plt.title('Average Price by Body Style', fontsize=20)
plt.show()
```

The bar chart above displays how different body styles affect car prices as the body type with hardtop prices high and body type with hatchback prices less. We have also shown this by finding average price of cars grouped by their body types shown in pie chart.

### Visualize the relationship between body type and the number of cylinders

```python
#bar chart
plt.figure(figsize=(10, 6))
body_style_cylinders = df.groupby(['body-style', 'num-of-cylinders']).size().unstack(fill_value=0)
body_style_cylinders.plot(kind='barh', stacked=True, figsize=(10, 6))
plt.title('Body Style vs Number of Cylinders',fontsize=16)
plt.xlabel('Body Style',fontsize=12)
plt.ylabel('Number of Cars',fontsize=12)
plt.show()
```

Similar to the previous chart,This stacked bar chart showcases how the number of cylinders varies across different body styles. It provides insights into the predominant cylinder configurations within each body type, offering a view of engine diversity across body styles.
